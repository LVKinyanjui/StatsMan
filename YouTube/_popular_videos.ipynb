```python
import os
import pandas as pd
from pyautogui import prompt
from googleapiclient.discovery import build
```

```python
API_KEY = os.getenv("YOUTUBE_API_KEY")
youtube = build('youtube', 'v3', developerKey=API_KEY)
```

```python
# Define the region code for the desired region (e.g., 'US' for the United States)
region_code = 'KE'
```

```python
# Call the videos.list method to retrieve popular videos by region
popular_videos_response = youtube.videos().list(
    chart='mostPopular',
    regionCode=region_code,
    part='snippet,statistics',
    maxResults=10  # You can adjust the number of results you want
).execute()
```

```python
# Extract and print video details
contents = []
for video_result in popular_videos_response.get('items', []):
    video_id = video_result['id']
    video_title = video_result['snippet']['title']
    video_channel = video_result['snippet']['channelTitle']
    publish_date = video_result['snippet']['publishedAt']
    view_count = video_result['statistics']['viewCount'] if 'statistics' in video_result else 'N/A'
    video_description = video_result['snippet']['description']

    contents.append(
        {
            "video_id": video_id,
            "video_title": video_title,
            "video_channel": video_channel,
            "publish_date": publish_date,
            "view_count": view_count,
            "video_description": video_description,
        }
    )
    
    print(f'Video Title: {video_title}')
    print(f'Video ID: {video_id}')
    print(f'Channel: {video_channel}')
    print(f'Published Date: {publish_date}')
    print(f'View Count: {view_count}')
    print('--------------------------')
```

```python
data = pd.DataFrame(contents)
data.head()
```

```python
data.to_csv("data/popular_videos.csv")
```
